## Tugas Praktikkum

```python
import pandas as pd

pd.set_option('display.max_rows', 20)       # Maksimal baris tampil
pd.set_option('display.max_columns', 10)    # Maksimal kolom tampil
pd.set_option('display.width', 100)         # Lebar tampilan
pd.set_option('display.float_format', '{:.2f}'.format)  # Format angka desimal

df = pd.read_csv('./epl-goalScorer(20-21).csv')

print("5 Data Pertama:\n", df.head())

print("\nJumlah Baris dan Kolom:", df.shape)

print("\nTipe Data Kolom:\n", df.dtypes)

print("\nJumlah Goal Tiap Pemain:\n", df[['player_name', 'goals']])

print("\nKolom Tertentu:\n", df[['player_name', 'goals', 'position', 'team_title', 'yellow_cards']])

print("\nPemain Liverpool:\n", df[df['team_title'] == 'Liverpool'][['player_name', 'team_title']])

print("\nPemain Arsenal:\n", df[df['team_title'] == 'Arsenal'][['player_name', 'team_title']])

print("\n10 Kartu Kuning Terbanyak:\n", df[['player_name', 'yellow_cards']].sort_values(by='yellow_cards', ascending=False).head(10))

print("\n10 Kartu Merah Terbanyak:\n", df[['player_name', 'red_cards']].sort_values(by='red_cards', ascending=False).head(10))
```
